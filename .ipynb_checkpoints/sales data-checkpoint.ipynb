```python
from scipy import stats
```

```python
boxcox_overvalue,lambda_value = stats.boxcox(final_df['order_value_EUR'])
sns.kdeplot(boxcox_overvalue, fill= True)
```

```python
boxcox_cost, lambda_value = stats.boxcox(final_df['cost'])
sns.kdeplot(boxcox_cost, fill= True)
```

```python
final_df['Yeo Cost'] = yeo_cost
```

```python
final_df.head()
```

```python
PC_data = final_df[final_df['device_type'] == 'PC']
Tab_data = final_df[final_df['device_type'] == 'Tablet']
mob_data = final_df[final_df['device_type'] == 'Mobile']
```

```python
f_statistic, p_value = stats.f_oneway(PC_data['Yeo Cost'], Tab_data['Yeo Cost'], mob_data['Yeo Cost'])

print("One-Way ANOVA F=Statistic: ", f_statistic)
print("P-Value: ", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis. ")
```

```python
correlation1, p_value1 = stats.pearsonr(boxcox_overvalue,yeo_cost)
correlation2, p_value2 = stats.pearsonr(boxcox_overvalue,final_df['refund'])
correlation3, p_value3 = stats.pearsonr(yeo_cost,final_df['refund'])
```

```python

print("Pearson Correlation Coficient for Overvalue and Cost: ", correlation1)
print("P-Value: ", p_value1)

print("Pearson Correlation Coficient for Overvalue and Refund: ", correlation2)
print("P-Value: ", p_value2)

print("Pearson Correlation Coficient for Cost and Overvalue: ", correlation3)
print("P-Value: ", p_value3)

if p_value1 < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis. ")

if p_value2 < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis. ")

if p_value3 < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis. ")
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(np.column_stack((boxcox_overvalue, yeo_cost)))

model = sm.OLS(final_df['refund'], X).fit()

coeff = model.params

print("Regration Coeffecient")
print("Intercept: ", coeff[0])
print("Oder Value: ", coeff[1])
print("Cost: ", coeff[2])

print("\nRegration Summary: ")
model.summary()
```
